```python
from imageai.Detection import ObjectDetection
import os

#execution_path = os.getcwd()

detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath("/content/drive/MyDrive/ObjectDetection/resnet50_coco_best_v2.1.0.h5")
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image= "/content/drive/MyDrive/ObjectDetection/0000002_00005_d_0000014.jpg", output_image_path= "/content/drive/MyDrive/ObjectDetection/newimage.jpg")

for eachObject in detections:
    print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
```
